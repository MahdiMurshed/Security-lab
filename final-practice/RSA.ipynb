# 1. RSA Encryption decryption

```python
def textToNum(t):
    nums = [ord(c) -  base + 1 for c in t]
    return nums
```

```python
def numsToTxt(arr):
    txt = "".join(chr(it + base -1) for it in arr)
    return txt
```

```python
def modPow(t, n, e):
    if e == 0:
        return 1
    res = modPow(t, n, e//2)
    res = (res*res) % n

    if e % 2 == 1:
        res = (res * t) % n
    return res
```

```python
def enc(t, n, e):
    ci = [modPow(c, n, e) for c in t]
    return ci
```

```python
def dec(t, n, d):
    ci = [modPow(c, n, d) for c in t]
    return ci
```

# 2. Calculate d

```python
def exEu(a, b):
    if b == 0:
        return a, 1,0
    gcd, x1, y1 = exEu(b, a%b)
    x = y1
    y = x1 - (a//b) * y1

    return gcd, x, y
```

```python
def modInv(e, phi_n):
    gcd, x, y = exEu(e, phi_n)

    if gcd != 1:
        print("Inverse n/A")
        return
    x = (x % phi_n + phi_n) % phi_n
    return x
```

# 4. RSA signature

```python
signature_file = "signature.bin"
text_file = "text.txt"
key_file = "key.bin"
```

```python
from hashlib import sha512
import pickle
import os
```

```python
if not os.path.exists(key_file):
    keyPair = RSA.generate(bits = 1024)
    key = {
    'n': keyPair.n,
    'e': keyPair.e,
    'd': keyPair.d
    }
    with open(key_file, 'wb') as file:
        pickle.dump(key, file)
else:
    with open(key_file, 'rb') as file:
        key = pickle.load(file)    
```

```python
with open(text_file, 'rb') as file:
    pt = file.read()
```

```python
hash = int.from_bytes(sha512(pt).digest(), byteorder="big")
sig = pow(hash, key['d'], key['n'])
sig
```

```python
with open(signature_file, 'wb') as file:
    pickle.dump(sig, file)
```
