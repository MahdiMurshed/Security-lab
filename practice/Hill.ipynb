```python
import numpy as np
```

```python
def generateKeyMatrix(key):
    keyMatrix = np.zeros((3, 3), dtype=int)
    index = 0

    for i in range(3):
        for j in range(3):
            keyMatrix[i, j] = ord(key[index]) - ord('a')
            index += 1

    return keyMatrix
```

```python
def generateMessageMatrix(message):
    messageMatrix = np.zeros((3, 196), dtype = int)
    index =0

    for i in range(196):
        for j in range(3):
            messageMatrix[j, i] = ord(message[index]) - ord('a')
            index += 1
    return messageMatrix
```

```python
# Example usage:
# Assuming you have the plainMatrix and keyMatrix already defined as NumPy arrays
# plainMatrix = np.array([[...], [...], [...]])  # Shape: (3, 196)
# keyMatrix = np.array([[...], [...], [...]])    # Shape: (3, 3)
# encryptedMatrix = encrypt(plainMatrix, keyMatrix)
def encrypt(plainMatrix, keyMatrix):
    encryptedMatrix = np.zeros((3, 196), dtype=int)

    for i in range(3):
        for j in range(196):
            encryptedMatrix[i, j] = np.sum(keyMatrix[i] * plainMatrix[:, j]) % 26

    return encryptedMatrix
```
